Seaborn library used for data visualization

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#total bill vs tip dataset it is inbuilt in seaborn
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a theme
sns.set_theme()
```

```python
#visualizing the dataset
sns.relplot(x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size', data=tips)
```

```python
#load the iris dataset
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
#scatter plot
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```

```python
#titanic dataset
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
#count plot
sns.countplot(x='class',hue='who',data=titanic)
```

```python
sns.countplot(x='class',hue='survived',data=titanic)
```

```python
#bar chart
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
#house price dataset
from sklearn.datasets import fetch_california_housing
house=fetch_california_housing()
house_df=pd.DataFrame(house.data,columns=house.feature_names) #converts numpy array house into datafram house_df
house_df['price']=house.target
```

```python
house_df.head()
```

```python
#distribution plot
sns.displot(house_df['price'])
```

```python
sns.distplot(house_df['price'])
```

```python
#correlation: positive if one value is dire ctly proportional
#negative if one value is inversely proportional
correlation=  house_df.corr()
```

```python
#heat map
plt.figure(figsize=(10,10)) #10 rows 10 columns
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
#cbar gives the colour bar on the side
#square gives the shape of each
#fmt .1f gives 1 decimal point
```
