MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
print(x)
print(y)
print(z)
```

```python
#plotting the data
#sin wave
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.show()
```

```python
#parabola
x=np.linspace(-5,5,100)
y=x**2 #y is x square
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('x^2')
plt.show()

```

```python
#plotting the parabola with + sign
plt.plot(x,y,'g+') #g means green
plt.xlabel('x')
plt.ylabel('x^2')
plt.show()
```

```python
#plotting both sin and cos in single plot in different colours
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,'g-')
plt.plot(x,z,'r--')
plt.xlabel('x')
plt.ylabel('sin(x) and cos(x)')
plt.legend()
plt.show()
```

```python
#bar plot
fig=plt.figure()
ax=fig.add_axes([0,0,1,1]) #0,0n is the origin and 1,1 is the width and height of the bar
langs=['c','c++','java','python','php']
students=[23,17,35,29,12]
ax.bar(langs,students)
plt.xlabel('languages')
plt.ylabel('number of students')
plt.show
```

```python
#pie chart
fig1=plt.figure()
ax1=fig1.add_axes([0,0,1,1])
langs=['c','c++','java','python','php']
students=[23,17,35,29,12]
ax1.pie(students,labels=langs,autopct='%1.2f%%') #autopct is the percentage of the pie chart
plt.show()
```

```python
#scatter plot
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([5,2,4,2,1,4,5,2])
plt.scatter(x,y)
plt.show()
```

```python
#scatter plot of sin and cosin
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.scatter(x,y,color='g')
plt.scatter(x,z,color='r')
```
