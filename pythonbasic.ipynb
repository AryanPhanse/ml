```python
#list it is enclosed in [], it can have duplicate values
my_list=[1,2,3,4]
my_list2=[1,"Hello",3.4]
print(my_list)
print(my_list2)
#add element to list
my_list2.append(5)
print(my_list2)
#print a speicific elemrnt in list
print(my_list[1])
#print length
print(len(my_list))
#delete an item in list2
del(my_list2[2])
print(my_list2)
#join 2 lists
print(my_list+my_list2)
```

```python
#Tuple it is immutable ie you cannot add or delete elements
tuple_1=(2,3,4,5)
print(tuple_1)
tuple_2=(2,"Hello",3.4)
print(tuple_2)
#converting list to tuple
my_list=[1,2,3,4]
print(tuple(my_list))
print(tuple_1[1])
#length of tuple
print(len(tuple_1))
```

```python
#set no duplicate
my_set={1,2,3,4,5,5}
print(my_set)
#convert list to set
my_list=[1,2,3,4,5,5]
print(set(my_list))
```

```python
#dictionary it has key-value pair it does not allow duplicate values
my_dictionary={'name':'aryan','age':30,'country':'India'}
print(my_dictionary)
print(my_dictionary['name'])
print(my_dictionary.get('age'))
```

OPERATORS

```python
#EXPONENT
a=2
b=5
print(a**b)
mod=a%b
print(mod)
#comparison
print(a==b)
print(a!=b)
print(a>b)
print(a<b)
print(a>=b)
print(a<=b)
#logical
a=True
b=False
print(a and b)
print(a or b)
print(not a)
##membership operators IN , Not in
my_list=[1,2,3,4,5]
print(3 in my_list)
print(6 in my_list)
print(6 not in my_list)
```

```python
#functions
#factorial
def factorial(n):
  if n==0:
    return 1
  else:
    return n*factorial(n-1)
print(factorial(int(input("Enter number"))))
```
